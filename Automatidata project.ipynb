```python
# Import libraries and packages
import numpy as np
import pandas as pd

# Load dataset into dataframe
df = pd.read_csv('C2_2017_Yellow_Taxi_Trip_Data')
print("done")
```

```python
# Sort the data by trip distance from maximum to minimum
df_sort = df.sort_values(by=['trip_distance'], ascending=False)
df_sort.head(10)
```

```python
# Sort the data by total amount and print the top 20 values
total_amount_sorted = df.sort_values(['total_amount'], ascending=False)['total_amount']
total_amount_sorted.head(20)
```

```python
# Print the bottom 20 values of the sorted total amount
total_amount_sorted.tail(20)
```

```python
df['payment_type'].value_counts()
```

```python
# Calculate the average tip for trips paid by credit card (type 1)
avg_cc_tip = df[df['payment_type'] == 1]['tip_amount'].mean()
print('Avg. cc tip:', avg_cc_tip)

# Calculate the average tip for trips paid by cash (type 2)
avg_cash_tip = df[df['payment_type'] == 2]['tip_amount'].mean()
print('avg_cash_tip:', avg_cash_tip)
```

```python
df['VendorID'].value_counts()
```

```python
df.groupby(['VendorID']).mean(numeric_only=True)[['total_amount']]
```

```python
# Filter the data for credit card payments only
data_credit_card = df[df['payment_type'] == 1]

# Count the frequency of each passenger count
data_credit_card['passenger_count'].value_counts()
```

```python
data_credit_card.groupby(['passenger_count']).mean(numeric_only=True)[['tip_amount']]
```
